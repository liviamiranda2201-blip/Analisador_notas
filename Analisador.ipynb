# 📊 Analisador de Notas

Este projeto é um programa em Python que analisa notas de alunos.  
Ele permite que o usuário digite várias notas e mostra:

1. A **média** das notas.  
2. Quais notas estão **acima da média**.  

---

## 🔹 Como funciona

1. O usuário digita as notas separadas por espaço.  
   Exemplo: `5 7 10 3 8`  

2. O programa transforma a entrada em uma **lista de números**.  
   - Cada número digitado é convertido de string para float.  

3. Calcula a **média** da lista.  

4. Percorre a lista e seleciona os números que são **maiores que a média**.  

5. Mostra o resultado na tela.  

---

## 🔹 Conceitos usados

- **Listas**: para armazenar as notas.  
- **Funções**: para organizar o código (ex: função `acima_media`).  
- **Loops (`for`)**: para percorrer cada nota.  
- **Condicional (`if`)**: para verificar se a nota está acima da média.  
- **Entrada do usuário (`input`)**: para receber os valores.  
- **Métodos de strings (`split()`)**: para separar os números digitados.  
- **Conversão de tipos (`float`)**: para transformar as strings em números.  

---

## 🔹 Exemplo de uso


```python
def acima_media(lista):
    media = sum(lista) / len(lista)
    resultado = []

    for valor in lista:
        if valor > media:
            resultado.append(valor)

    return resultado

#Entrada do usuário
entrada = input("Digite as notas separadas por espaço: ")

# transformar em lista de números
notas = [float(x) for x in entrada.split()]

# calcular média
media = sum(notas) / len(notas)

print("Média:", media)
print("Acima da média:", acima_media(notas))
```
